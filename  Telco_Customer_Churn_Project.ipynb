## **Question   ➊➑ :**
##We are planning to provide offers based on monthly charge categories; could you elaborate on that?

````python
min_value = df['MonthlyCharges'].min()
max_value = df['MonthlyCharges'].max()
median_value = df['MonthlyCharges'].median()
````

````python
print (f"min = {min_value}")
print (f"max = {max_value}")
print (f"median = {median_value}")
````

````python
def categorize_charges(charge):
 if charge < 50:
  return 'lower'
 elif charge <= 90:
  return 'median'
 else:
  return 'upper'

df['ChargeCategory'] = df['MonthlyCharges'].apply(categorize_charges)
category_counts = df['ChargeCategory'].value_counts()
category_counts
````

````python
plt.figure(figsize=(8, 6))
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%')
````

````python
def assign_offer_01(charge_category):

 if charge_category == 'lower':
  return '10% off for 6-month auto-renewal.'
 elif charge_category == 'median':
  return '1 month free of premium service.'
 elif charge_category == 'upper':
  return 'Free subscription to exclusive services.'
 else:
  return 'No offer'

df['offers_01'] = df['ChargeCategory'].apply(assign_offer_01)
````

````python
def assign_offer_02(row):

 gender = row['gender']
 partner = row['Partner']

 if gender == 'Male' and partner == 'Yes':
  return '10% off family subscriptions.'

 elif gender == 'Male' and partner == 'No':
  return 'One free month with Fiber optic subscription.'

 elif gender == 'Female' and partner == 'Yes':
  return 'One free month in StreamingMovies with StreamingTV subscription.'

 elif gender == 'Female' and partner == 'No':
  return 'One free month of premium service.'

 else:
  return 'No offer'

df['offers_02'] = df.apply(assign_offer_02, axis=1)
````
